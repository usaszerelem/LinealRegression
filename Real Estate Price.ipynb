```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
x = df['area']
y = df['price']

# Reshape array to look identical to the data. One column.
X = np.array(x).reshape(-1, 1)
X
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color="red", marker='+')
plt.plot(df.area, model.predict(df[['area']].values), color='blue')
```

```python
model.predict(np.array([3300]).reshape(-1, 1))
# model.predict(3300)
```

```python
# Internal details. The way this model predicts (calculates) the coeficient (m) and intercept (b)
# price = m * area + b
# Now that this two values are known, then the area (x) needs to be added for the prediction.
prediction = model.coef_ * 3300 + model.intercept_
prediction
```
